# Concat and Merge

```python
import  pandas as pd
concat1 = pd.read_csv('7-concat_data1.csv')
concat2 = pd.read_csv('7-concat_data2.csv')
merge1 = pd.read_csv('7-merge_data1.csv')
merge2 = pd.read_csv('7-merge_data2.csv')
```

```python
concat1
```

```python
concat2
```

```python
merge1
```

```python
merge2
```

```python
df_concat = pd.concat([concat1, concat2],ignore_index=True)
```

```python
df_concat
```

```python
#merge outer join
df_merge = pd.merge(merge1, merge2, on='Employee_ID',how='outer')
df_merge
```

```python
#left join
df_merge = pd.merge(merge1, merge2, on='Employee_ID',how='left')
df_merge
```

```python
#right join
df_merge = pd.merge(merge1, merge2, on='Employee_ID',how='right')
df_merge
```

```python
#inner join
df_merge = pd.merge(merge1, merge2, on='Employee_ID',how='inner')
df_merge
```

```python
def salary_category (salary):
    if salary < 50000:
        return "Low"
    elif 50000 < salary < 80000:
        return "Medium"
    else:
        return "High"
```

```python
df_apply['Salery_Category']=df_apply['Salary'].apply(salary_category)
df_apply
```

```python
def performanseScore(experience):
    if experience > 10:
        return 1
    else:
        return 0
```

```python
df_apply['Performance_Score']=df_apply['Performance_Score'] + df_apply['Experience'].apply(performanseScore)
df_apply
```
